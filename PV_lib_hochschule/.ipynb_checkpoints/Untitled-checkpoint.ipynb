```python
# -*- coding: utf-8 -*-
"""
Created on Sun Nov 27 18:43:02 2022

@author: Gaurang
"""

! pip install pvlib

import pvlib
import pandas as pd
import matplotlib.pyplot as plt
from pvlib.modelchain import ModelChain
from pvlib.location import Location
from pvlib.pvsystem import PVSystem
from pvlib.temperature import TEMPERATURE_MODEL_PARAMETERS
%matplotlib inline

sandia_modules=pvlib.pvsystem.retrieve_sam('SandiaMod')
cec_inverters=pvlib.pvsystem.retrieve_sam('CECInverter')

module=sandia_modules['Canadian_Solar_CS5P_220M___2009_']
inverter=cec_inverters['Huawei_Technologies_Co___Ltd___SUN2000_100KTL_USH0__800V_']
temperature_parameters=TEMPERATURE_MODEL_PARAMETERS['sapm']['open_rack_glass_glass']

location=Location(latitude=51.49524367854366,longitude=10.808466217221286,
                  tz='Europe/Berlin',altitude=189,name='Building_19')

system=PVSystem(surface_tilt=30,surface_azimuth=193,module_parameters=module,
                inverter_parameters=inverter,
                temperature_model_parameters=temperature_parameters,
                modules_per_string=7,strings_per_inverter=30)

location1=Location(latitude=51.49524367854366,longitude=10.808466217221286,
                  tz='Europe/Berlin',altitude=190,name='Building_20')
system1=PVSystem(surface_tilt=45,surface_azimuth=193,module_parameters=module,
                 inverter_parameters=inverter,
                 temperature_model_parameters=temperature_parameters,
                 modules_per_string=7,strings_per_inverter=30)

location2=Location(latitude=51.49524367854366,longitude=10.808466217221286,
                  tz='Europe/Berlin',altitude=180,name='Aki')
system2=PVSystem(surface_tilt=15,surface_azimuth=101,module_parameters=module,
                 inverter_parameters=inverter,
                 temperature_model_parameters=temperature_parameters,
                 modules_per_string=7,strings_per_inverter=30)

location3=Location(latitude=51.49524367854366,longitude=10.808466217221286,
                  tz='Europe/Berlin',altitude=180,name='In.Ret')
system3=PVSystem(surface_tilt=30,surface_azimuth=192,module_parameters=module,
                 inverter_parameters=inverter,
                 temperature_model_parameters=temperature_parameters,
                 modules_per_string=7,strings_per_inverter=30)

location4=Location(latitude=51.49524367854366,longitude=10.808466217221286,
                  tz='Europe/Berlin',altitude=202,name='Studienkolleg')
system4=PVSystem(surface_tilt=30,surface_azimuth=184,module_parameters=module,
                 inverter_parameters=inverter,
                 temperature_model_parameters=temperature_parameters,
                 modules_per_string=7,strings_per_inverter=30)

modelchain=ModelChain(system,location)
modelchain1=ModelChain(system1,location1)
modelchain2=ModelChain(system2,location2)
modelchain3=ModelChain(system3,location3)
modelchain4=ModelChain(system4,location4)

times1=pd.date_range(start="2021-01-01 00:00",end="2021-12-31 00:00",freq='1min',tz=location.tz)
clear_sky1=location1.get_clearsky(times1)
clear_sky1.resample('m').sum().plot(figsize=(16,9))

plt.subplots(figsize=(12,10),dpi=200)
plt.subplot(2,3,1)
modelchain.run_model(clear_sky1)
modelchain.results.ac.resample('m').sum().plot()
plt.xlabel('Month')
plt.ylabel('Energy yield')
plt.title('Energy yield of a rooftop \n installation of Building 19 ')
plt.subplot(2,3,2)
modelchain1.run_model(clear_sky1)
modelchain1.results.ac.resample('m').sum().plot()
plt.xlabel('Month')
plt.ylabel('Energy yield')
plt.title('Energy yield of a rooftop \n installation of Building 20 ')
plt.subplot(2,3,3)
modelchain2.run_model(clear_sky1)
modelchain2.results.ac.resample('m').sum().plot()
plt.xlabel('Month')
plt.ylabel('Energy yield')
plt.title('Energy yield of a rooftop \n installation of AKI ')
plt.subplot(2,3,4)
modelchain3.run_model(clear_sky1)
modelchain3.results.ac.resample('m').sum().plot()
plt.xlabel('Month')
plt.ylabel('Energy yield')
plt.title('Energy yield of a rooftop \n installation of In.Ret ')
plt.savefig('my_plot_10.png',bbox_inches='tight')
plt.subplot(2,3,5)
modelchain4.run_model(clear_sky1)
modelchain4.results.ac.resample('m').sum().plot()
plt.xlabel('Month')
plt.ylabel('Energy yield')
plt.title('Energy yield of a rooftop \n installation of Studienkolleg ')
plt.tight_layout()   

plt.show()
    
```
